<h1>Abortion Rates in America</h1>

<h2>Introduction</h2>

Since the landmark Supreme Court decision of Roe v. Wade in 1973, the right to abortion has been protected in the United States at the federal level. However, in a country as large and diverse as America, opinions on abortion have always varied greatly between states. Until recently, states were unable to ban abortion outright. They were allowed to restrict the procedure in other ways, such as blocking funding for abortion providers or banning it after a certain amount of time. One would expect states that implemented such laws to have lower rates of abortion. However, other factors such as availability of contraception and access to comprehensive sex education might skew the relationship. Broadly, the states that tend to de-emphasize contraception and sex education are the ones who tend to restrict abortion. Thus, any decrease in the number of abortions in the state due do lack of access might be gained back because more people, particularly young people, who have little knowledge of safe sex practices and few resources to properly support a child, are getting pregnant and seeking an abortion out. This pattern of thought is generally found in more conservative, or “red” states, whereas more liberal (“blue”) states have more access to abortion and lower rates of teen pregnancy. However, almost 60% of the people who receive abortions in the United States already have at least one child – adolescents, especially in more recent years, make up only a small fraction of all abortions (via <a href="https://www.guttmacher.org/united-states/abortion">Guttmacher Institute</a>). So how big of an effect would lack of sex education and increased teen pregnancy really have on abortion overall? It is a complicated question to be sure.


<b>Research Question:</b> Does the political affiliation of a state influence that state's rate of abortion?

<b>Hypothesis:</b> Abortion rates will be lower in red states due to state laws that limit access to the procedure

Typically a "Red" state is a state that leans towards Republican values, while a "Blue" state is a state that leans towards Democratic values. 

Republican values align with a more conservative approach in policy. Conservative approaches aim to maintain or restore moral values and coincide with a more religious viewpoint on issues. Namely, abortion is viewed as an end-of-life method that is shunned as immoral. The choice to end a term is viewed as not being one that the individual should make. 

Democratic values align with a more liberal approach in policy. Liberal apporaches aim to maintain or progress humanistic values and coincide with individual liberties. They tend to deal with decision-making being at the hands on the individual and ths abortion is seen as act that should be allowable for the carrier of the child. The choice to end a term is viewed as being first and foremost the choice of the mother. 

The crux of our research is to see which states, Red or Blue has higher abortion rates. In addition, a narrower set of findings may develop where we can see the abortion rates in these states amongst different age groups. 

**Considering What is Red and What is Blue**

Distinguishing between a red and blue state is multi-layered and complex yet nonetheless the distinction must be made. The passage of legislation deals with various branches of the government but on a state and local level. This also includes a "lag" in the implementation of laws, whether they be red-leaning or blue-leaning.

##Fix outliers 

```python
dfCopy = df.copy()

#method for finding the limits for detecting outliers. 
def limits(att):
  a, b = np.nanpercentile(att, [75,25])
  c = a-b
  upperlimit = a + 1.5*c
  lowerlimit = b - 1.5*c
  return upperlimit,lowerlimit

df.drop('state',inplace = True, axis =1)
df.drop('stateColor',inplace=True, axis =1)

for i in df:
  upperlimit,lowerlimit = limits(df[i])
  df[i]= np.where((df[i]>upperlimit)|(df[i]<lowerlimit),np.nan, df[i])
  df[i].fillna(df[i].median(), inplace =True)

df['state'] =  dfCopy['state']
df['stateColor']= dfCopy['stateColor']
 
```

##Elastic net regression

Elastic net regression is best used for data that have many independent features and especially if those features are correlated to each other. The elastic net regression is a combination of both lasso and ridge regression. If we were to just use ridge regression for such dataset, it will shrink all the coefficients for the features but the model will stay be complex. Also, if we were to just use lasso regression for such dataset, it would only keep one variable while setting all other coefficient to 0, which would lead to lower accuracy in our model. Elastic net regression has a balance of both penalties, which improves the performance and can result in better model vs if modeled individually.

#Ensemble Modeling 

Ensemble modeling helps reduce overfitting and high variance in our model. How it works is, we generate n number of different models and we use the models to predict the dependent variable of the dataset. For our dataset, we will use the bagging method.

Bagging method:
* In the bagging method, when we train our models, our train data will be generated by resampling with replacement so the samples values would be independent.

* Then we predict the y value

* The result we get will be the averge of all the different model predictions. 

In this way, even if individually each model is weak, when we combine the results. You get overall a good quality result.

For our dataset we will use random forest algorithm. Random forest algorithm is one of the bagging techniques, with only one difference. Which is when we split the data into training and testing data. We not only sample the data rows, but also columns. One model could have training data of 3 column, while another could have a training data of 5 columns. Then the rest would be the same, we would train and test. Then with the mean value from all the models we would predict the values. 




```python
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor

#Copy scaled data 
tempDf = dfScaled.copy()

#Assign X and y 
y = tempDf['abortionratetotal']
tempDf.drop(['abortionratetotal'], axis=1, inplace=True)
X = tempDf

#Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = .20, random_state = 10)
```

```python

#n_estimators is the number of trees to be used in the forest. 
dt = RandomForestRegressor(n_estimators=100, random_state= 10)  

#set up bagging regressor, with the base_estimator which is technique as RandomForestRegressor, and the sample are drawn with replacement.
bc = BaggingRegressor(base_estimator = dt, bootstrap = True)

#fit the data
dt.fit(X_train, y_train)

#Predict the data
y_preds = dt.predict(X_test)
```

# Conclusion

Our original hypothesis stated that abortion rates will be lower in red states due to state laws that limit access to the procedure. However, the data we have analyzed cannot definitively support this hypothesis. While some of our regression models are highly accurate, the issue is that abortion rate is highly correlated to other factors like teen abortion rate and pregnancy rate. Therefore, even with such a weak correlation between abortion rate and party affiliation (state color), the model is still able to perform well. In the correlation matrix, we can better see the truth of the matter: the r value between total abortion rate and state color is -0.17 (with the Pearson method). While a negative value does lend some credence to our hypothesis - conservative states that have the "higher" state color value (a 1 vs the 0 assigned to liberal states) would be expected to have a lower rate of abortion, and thus a negative correlation - the number is simply too close to 0 to be statistically relevant.

There could be many reasons for the lack of correlation. One explaination is that state abortion laws do not have as much of an effect on abortion rates as one might expect. Because Roe v Wade was in effect during all of the years used in the dataset, we know that despite whatever restrictions a state attempted to put on abortion, they would not have been able to ban the procedure entirely, lessening the impact that conservative abortion laws could potentially have on the abortion rate. Another explaination could be that there are other factors that influence the rate of abortion in each state. While we cannot confirm this hypothesis with data from our source, it is alost assuredly the case. Factors like avalability of contraception and robust sex education programs likely lessen the abortion rate, as these things aim to prevent unwanted pregnancy in the first place. However, paradoxically, conservative states that aim to limit abortion also generally have poor sex education and access to contraception, which would increase the need for abortion rather than decrease it. This fact would weaken any correlation we could have found.

Another reason could be the method by which we chose classify each state as "red" or "blue". If a state had conservative governor for the vast majority of the years we included, it was considered a red state. If the majority of the years were with a liberal governor, then it was considered blue. While we could have just gone year by year - if the governor was a Republican for that year then make the state red for that year, and vice versa. However, the reality of politics is that even if a governor is in office, that doesn't automatically mean they will be able to pass whatever abortion legislation they want as soon as they get to office. The state legislatures are the ones who create the laws - the governor merely signs them, and even then she or he can be overrulled with enough votes from the legislature. There is no guarantee that the governor and the state legislature have the same political affiliation. In addition, even if a governor would support abortion restrictions, for the most part getting the laws written, passed and them implemented is a difficult and timely task. Even if a sweeping limit was eventually made to abortion (which was not likely given the protection by Roe v Wade at the time), it would generally take years for it to make a difference in abortion numbers, if any. There is also the fact that abortion restrictions in one state can be skirted by traveling to another state.

All in all, there are simply too many factors than cannot be accounted for in our data to make the statement that conservative states have lower rates of abortion than liberal states. Abortion is too complicated and important an issue to be explained away by only party affiliation. Especially now, it is a matter of great importance that requires further data and interpretation.
